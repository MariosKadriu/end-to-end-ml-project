#### Correlations

```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# create an empty DataFrame to store the correlation coefficients and p-values
correlation_df = pd.DataFrame(columns=['numerical_col', 'categorical_col', 'corr_coeff', 'p_value'])

# calculate the correlation between each numerical and categorical column and store the results in the correlation_df DataFrame
for num_col in numeric_features:
    for cat_col in categorical_features:
        corr, p_value = stats.pointbiserialr(df[num_col], pd.get_dummies(df[cat_col])[df[cat_col].unique()[0]])
        correlation_df = correlation_df.append({'numerical_col': num_col, 'categorical_col': cat_col, 'corr_coeff': corr, 'p_value': p_value}, ignore_index=True)

# create a pivot table from the correlation_df DataFrame
pivot_table = pd.pivot_table(correlation_df, values='corr_coeff', index='numerical_col', columns='categorical_col')

# create a heatmap plot using Seaborn
heatmap = sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=11)
heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=11)

plt.show()
```
